```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
movieData = pd.read_csv("movies.csv")
movieData.head()
```

```python
ratingData = pd.read_csv("ratings.csv")
ratingData.head()
```

```python
allData = pd.merge(movieData, ratingData, on="movieId")
allData.head()
```

```python
allData.groupby("title")["rating"].mean().sort_values(ascending=False).head(10)
```

```python
allData.groupby("title")["rating"].count().sort_values(ascending=False).head(10)
```

```python
ratMeanCount = pd.DataFrame(allData.groupby("title")["rating"].mean())
ratMeanCount
```

```python
ratMeanCount["rating count"] = allData.groupby("title")["rating"].count()
ratMeanCount.head()
```

```python
for i in range(ratMeanCount["rating"].shape[0]):
    ratMeanCount["rating"][i] = round(ratMeanCount["rating"][i],1)

ratMeanCount.head()
```

```python
ratMeanCount = ratMeanCount[(ratMeanCount["rating"] > 3) & (ratMeanCount["rating count"] > 100)]
ratMeanCount
```

```python
ratMeanCount = ratMeanCount.sort_values(by="rating", ascending=False).head(10)
ratMeanCount
```
